# Tensorflow Data Validation Testing

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load Data and Process
train = pd.read_csv("data/train.csv")
private_test = pd.read_csv("data/test.csv")
```

### Load TFDV

```python
import tensorflow_data_validation as tfdv
```

```python
stats_train = tfdv.generate_statistics_from_dataframe(train)
```

```python
tfdv.visualize_statistics(stats_train)
```

```python
schema = tfdv.infer_schema(stats_train)
```

```python
tfdv.display_schema(schema)
```

```python
# Lets try the anomaly detect functions
stats_test = tfdv.generate_statistics_from_dataframe(private_test)
```

```python
test_anomalies = tfdv.validate_statistics(stats_test, schema)
```

```python
tfdv.display_anomalies(test_anomalies)
```

```python
# Add the schemas
schema.default_environment.append('TRAIN')
schema.default_environment.append('TEST')
```

```python
tfdv.get_feature(schema, 'target').not_in_environment.append('TEST')
```

```python
# recheck the schema info
recheck_test_anomalies = tfdv.validate_statistics(stats_test, schema, environment='TEST')
```

```python
tfdv.display_anomalies(recheck_test_anomalies)
```

## Comparing Datasets

```python
tfdv.visualize_statistics(lhs_statistics=stats_train, rhs_statistics=stats_test, lhs_name='TRAIN', rhs_name='TEST')
```
